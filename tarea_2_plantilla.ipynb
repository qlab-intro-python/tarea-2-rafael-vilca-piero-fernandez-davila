12. Implement your fix to the major problem. Remake the plot from the previous question.

```python
import pandas as pd
import numpy as np

# Importar la base de datos
datos = pd.read_csv("C:/Users/ravir/OneDrive/Documentos/Git hub/Tarea 2/tarea-2-rafael-vilca-piero-fernandez-davila/flights.csv")
vuelos = pd.DataFrame(datos)

# Eliminar filas con NaN en 'arr_time' o 'dep_time'
vuelos = vuelos.dropna(subset=['arr_time', 'dep_time'])

# Convertimos 'dep_time' y 'arr_time' a minutos desde la medianoche
vuelos['dep_time_min'] = vuelos.dep_time // 100 * 60 + vuelos['dep_time'] % 100
vuelos['arr_time_min'] = vuelos.arr_time // 100 * 60 + vuelos['arr_time'] % 100

# Calcular la situación en la que los vuelos son de un día para otro, para evitar air_time0 negativos. En caso arr_time_min sea menor a dep_time_min, entonces se le sumará 1440 al resultado, de modo que se pueda obtener el tiempo efectivo de viaje. 

vuelos.air_time0 = np.where(
    vuelos['arr_time_min'] >= vuelos['dep_time_min'],
    vuelos['arr_time_min'] - vuelos['dep_time_min'],
    vuelos['arr_time_min'] - vuelos['dep_time_min'] + 1440
)

vuelos

plt.plot(vuelos.air_time, vuelos.air_time0, 'o')

#A pesar de haber solucionado los dos problemas mencionados, vemos que air_time no es igual a air_time0, lo cual se puede notar en el gráfico.

```

14. Calculate average delays by destination for flights originating in NYC. Create a variable which ranks destinations from worst to best, where 1 is the worst rank.

20. Write a code which calculates your delay measures separately by airline. Do particular airlines perform poorly on multiple metrics?

```python
# Solution


#Promedio de retraso en la llegada por aerolínea
promedio_por_aerolinea = vuelos.groupby(['carrier'])['arr_delay'].mean()

#Desviación estándar del retraso en la llegada por aerolínea
desviacion_por_aerolinea = vuelos.groupby(['carrier'])['arr_delay'].std()

#Porcentaje de vuelos 
vuelos_puntuales_por_aerolinea = vuelos[vuelos['arr_delay'] <= 0].groupby('carrier').size()
total_vuelos_por_aerolinea = vuelos.groupby(['carrier']).size()
porcentaje_puntuales = (vuelos_puntuales_por_aerolinea / total_vuelos_por_aerolinea) * 100


promedio_df = pd.DataFrame({'Promedio retraso': promedio_por_aerolinea})
desviacion_df = pd.DataFrame({'Desviación estándar retraso': desviacion_por_aerolinea})
porcentaje_df = pd.DataFrame({'Ratios llegadas puntuales': porcentaje_puntuales})

resultados = pd.concat([promedio_df, desviacion_df, porcentaje_df], axis=1)

resultados



#El caso más resaltante es el de la aerolinea FL, cuyas metricas en los 3 indicadores es pobre. Ya que su promedio de minutos de tardanza es el 2do mayor, como también tiene una desviación estandar alta y un porcentaje de llegadas puntuales baja en comparación al resto.
```
